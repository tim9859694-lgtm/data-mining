```python
import pandas as pd
# df = pd.read_csv("members.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch2/members.csv")
# 1) 표준편차, 평균값 구하기
std = df['age'].std() #데이터 프레임의 표준편차를 구하는 함수
mean = df['age'].mean() #데이터 프레임의 평균을 구하는 함수
# 2) 이상치 최저, 최고 기준 구하기
lower = mean - (std * 1.5) #평균에서 표준편차의 1.5배 만큼 뺀 값을 이상치 최저
upper = mean + (std * 1.5) #평균에서 표준편차의 1.5배 만큼 더한 값을 이상치 최고로 설정하는 코드
# 3) 이상치를 벗어나는 값(조건) 찾기
cond1 = df['age'] < lower #lower값보다 작거나
cond2 = df['age'] > upper #upper 값보다 큰것을 이상치를 벗어나는 값으로 간주하는 코드

# 4) 조건에 만족하는 이상치 age합
print(df[cond1|cond2]['age'].sum())
```

```python
import pandas as pd
# df = pd.read_csv("members.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch2/members.csv")

# 1) 앞에서 부터 80% 데이터 슬라이싱
line = int(len(df) * 0.8)
df = df.iloc[:line]

# 2) 결측치 채우기 전 'f1' 컬럼 표준편차 구하기
std1 = df['f1'].std()

# 3) 중앙값으로 결측치 채우기
med=df['f1'].median()
df['f1'] = df['f1'].fillna(med)

# 4) 결측치를 채운 후 'f1' 컬럼 표준편차 구하기
std2 = df['f1'].std()

# 5) 두 표준편차 차이 절대값 계산 (반올림하여 소수 둘째자리까지 계산)
result = abs(std1-std2)
print(round(result,2))
```
