```python
from google.colab import drive
from pathlib import Path
import pandas as pd
from scipy import stats
import os

if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

FILE_PATH = '/content/drive/MyDrive/type3_week1/delivery_time.csv'

if os.path.exists(FILE_PATH):
    # CSV 불러오기
    df = pd.read_csv(FILE_PATH)

    # 분석 대상 (결측치 제거)
    x = df['delivery_min'].dropna()

    result = stats.ttest_1samp(x, popmean=30, alternative='less')

    # 통계량 계산
    mean_x = x.mean()
    n = len(x)
    se = stats.sem(x)
    # 95% 신뢰구간
    ci_low, ci_high = stats.t.interval(0.95, df=n-1, loc=mean_x, scale=se)

    # 결과 출력
    print('[데이터 미리보기]')
    print(df.head(), '\n')
    print('[분석 결과]')
    print(f'표본 수: {n}')
    print(f'표본평균: {round(mean_x, 4)}')
    print(f't-statistic: {round(result.statistic, 4)}')
    print(f'p-value: {round(result.pvalue, 6)}')
    print(f'95% 신뢰구간: ({round(ci_low, 4)}, {round(ci_high, 4)})')

    # 결론 출력 (유의수준 0.05 기준)
    if result.pvalue < 0.05:
        print("\n결론: p-value가 0.05보다 작으므로 귀무가설(H0)을 기각함. 평균 배달시간은 30분 미만이라고 볼 수 있음.")
    else:
        print("\n결론: p-value가 0.05보다 크므로 귀무가설(H0)을 채택함. 평균 배달시간이 30분 미만이라고 하기 어려움.")

else:
    print(f"오류: '{FILE_PATH}' 경로에 파일이 없음.")
    print("코랩 왼쪽 '폴더' 아이콘에서 파일 위치를 다시 확인하고 경로를 수정하셈.")
```
