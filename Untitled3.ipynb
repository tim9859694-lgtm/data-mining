```python
from google.colab import drive
from pathlib import Path
import pandas as pd
from scipy import stats
import os

if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

FILE_PATH = '/content/drive/MyDrive/type3_week1/delivery_time.csv'

if os.path.exists(FILE_PATH):
    # CSV 불러오기
    df = pd.read_csv(FILE_PATH)

    # 분석 대상 (결측치 제거)
    x = df['delivery_min'].dropna()

    result = stats.ttest_1samp(x, popmean=30, alternative='less')

    # 통계량 계산
    mean_x = x.mean()
    n = len(x)
    se = stats.sem(x)
    # 95% 신뢰구간
    ci_low, ci_high = stats.t.interval(0.95, df=n-1, loc=mean_x, scale=se)

    # 결과 출력
    print('[데이터 미리보기]')
    print(df.head(), '\n')
    print('[분석 결과]')
    print(f'표본 수: {n}')
    print(f'표본평균: {round(mean_x, 4)}')
    print(f't-statistic: {round(result.statistic, 4)}')
    print(f'p-value: {round(result.pvalue, 6)}')
    print(f'95% 신뢰구간: ({round(ci_low, 4)}, {round(ci_high, 4)})')

    # 결론 출력 (유의수준 0.05 기준)
    if result.pvalue < 0.05:
        print("\n결론: p-value가 0.05보다 작으므로 귀무가설(H0)을 기각함. 평균 배달시간은 30분 미만이라고 볼 수 있음.")
    else:
        print("\n결론: p-value가 0.05보다 크므로 귀무가설(H0)을 채택함. 평균 배달시간이 30분 미만이라고 하기 어려움.")

else:
    print(f"오류: '{FILE_PATH}' 경로에 파일이 없음.")
    print("코랩 왼쪽 '폴더' 아이콘에서 파일 위치를 다시 확인하고 경로를 수정하셈.")
```

```python
from google.colab import drive
import pandas as pd
from scipy import stats
from pathlib import Path
import os

# 1. 드라이브 마운트 (파일 읽기 전 필수)
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

# 2. 경로 설정 및 파일 로드
DATA_DIR = Path('/content/drive/MyDrive/type3_week1')
file_path = DATA_DIR / 'before_after_worktime.csv'

if file_path.exists():
    df = pd.read_csv(file_path)

    # 3. 데이터 확인
    print('[데이터 미리보기]')
    print(df.head(), '\n')

    # 4. 대응표본 t-test (Paired t-test)
    # H0: before - after == 0
    # H1: before - after > 0 (업무시간 감소)
    before = df['before_min']
    after = df['after_min']
    diff = before - after

    # ttest_rel은 (A, B, alternative='greater')일 때 A > B인지 확인함
    result = stats.ttest_rel(before, after, alternative='greater')

    # 5. 결과 출력
    print('[분석 결과]')
    print(f'표본 수: {len(df)}')
    print(f'평균 차이(before - after): {round(diff.mean(), 4)}')
    print(f't-statistic: {round(result.statistic, 4)}')
    print(f'p-value: {round(result.pvalue, 6)}')

    print('\n[결과 해석]')
    if result.pvalue < 0.05:
        print('p-value가 0.05보다 작으므로 교육 후 업무처리시간이 통계적으로 유의미하게 감소했다고 볼 수 있음.')
    else:
        print('p-value가 0.05보다 크므로 교육 후 업무처리시간이 감소했다고 보기 어려움.')
else:
    print(f"오류: '{file_path}' 경로를 확인하셈. 파일이 없음.")
```
